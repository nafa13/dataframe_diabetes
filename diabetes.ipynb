```python
df_bandung = df[df['nama_kabupaten_kota'] == 'KABUPATEN BANDUNG']

display(df_bandung)
```

```python
total_per_tahun = df.groupby('tahun')['jumlah_penderita_dm'].sum().reset_index()
display(total_per_tahun)
```

```python
rata_per_kab = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].mean().reset_index()
display(rata_per_kab.head())
```

```python
total_per_kab_all = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].sum()
kab_tertinggi = total_per_kab_all.idxmax()
nilai_tertinggi = total_per_kab_all.max()
kab_terendah = total_per_kab_all.idxmin()
nilai_terendah = total_per_kab_all.min()

print(f"Tertinggi: {kab_tertinggi} (Total: {nilai_tertinggi:,.0f})")
print(f"Terendah : {kab_terendah} (Total: {nilai_terendah:,.0f})")
```

```python
def kategori_dm(jumlah):
    if jumlah < 50000:
        return 'Rendah'
    elif 50000 <= jumlah <= 99999:
        return 'Sedang'
    else: 
        return 'Tinggi'
```

```python
df['kategori_dm'] = df['jumlah_penderita_dm'].apply(kategori_dm)
display(df[['nama_kabupaten_kota', 'jumlah_penderita_dm', 'kategori_dm']].head())
```

```python
total_thn = df.groupby('tahun')['jumlah_penderita_dm'].transform('sum')
df['persentase_tahun'] = (df['jumlah_penderita_dm'] / total_thn) * 100
display(df[['tahun', 'nama_kabupaten_kota', 'jumlah_penderita_dm', 'persentase_tahun']].head())
```

```python
tabel_ringkas = df.groupby('tahun').agg({
    'jumlah_penderita_dm': 'sum',      
    'nama_kabupaten_kota': 'count'     
}).reset_index()
```

```python
tabel_ringkas.columns = ['tahun', 'total_jumlah_penderita_dm', 'jumlah_kabupaten_kota']
display(tabel_ringkas)
```

```python
plt.figure(figsize=(15, 7))
plt.bar(df_2019['nama_kabupaten_kota'], df_2019['jumlah_penderita_dm'], color='teal')
plt.title('Jumlah Penderita DM per Kabupaten/Kota di Jawa Barat (2019)', fontsize=14, fontweight='bold')
plt.xlabel('Nama Kabupaten/Kota', fontsize=12)
plt.ylabel('Jumlah Penderita DM', fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(total_per_tahun['tahun'], total_per_tahun['jumlah_penderita_dm'],
         marker='o', linestyle='-', color='darkred', linewidth=2, markersize=8)
plt.title('Tren Total Jumlah Penderita DM Jawa Barat Per Tahun', fontsize=14, fontweight='bold')
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Total Jumlah Penderita DM', fontsize=12)
plt.xticks(total_per_tahun['tahun']) 
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_10 = df_2019.sort_values(by='jumlah_penderita_dm', ascending=False).head(10)
top_10_plot = top_10.sort_values(by='jumlah_penderita_dm', ascending=True)
plt.barh(top_10_plot['nama_kabupaten_kota'], top_10_plot['jumlah_penderita_dm'], color='salmon')
plt.title('10 Kabupaten/Kota dengan Penderita DM Tertinggi (2019)', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah Penderita DM', fontsize=12)
plt.ylabel('Kabupaten/Kota', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df_2019 = df[df['tahun'] == 2019]
proporsi = df_2019['kategori_dm'].value_counts()
colors = ['#ff9999', '#66b3ff', '#99ff99']
explode = (0.05, 0, 0)
plt.pie(proporsi, labels=proporsi.index, autopct='%1.1f%%', startangle=90,
        colors=colors, explode=explode if len(proporsi) == 3 else None, shadow=True)

plt.title('Proporsi Jumlah Kabupaten Berdasarkan Kategori DM (2019)', fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
tahun_terakhir = sorted(df['tahun'].unique())[-3:]
df_last_3 = total_per_tahun[total_per_tahun['tahun'].isin(tahun_terakhir)]
bars = plt.bar(df_last_3['tahun'].astype(str), df_last_3['jumlah_penderita_dm'], color='cornflowerblue', width=0.6)
plt.title(f'Perbandingan Total Penderita DM ({tahun_terakhir[0]}-{tahun_terakhir[-1]})', fontsize=14, fontweight='bold')
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Total Penderita', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{int(yval):,}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()
```
